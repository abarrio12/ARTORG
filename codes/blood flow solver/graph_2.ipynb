```python
import igraph as ig
import random
from igraph import layout
from pyvis.network import Network
```

```python
# Crear un grafo con 10 nodos y varias aristas
g2 = ig.Graph(10, directed = True)

# Añadir atributos a los nodos (nombre y color)
g2.vs["name"] = [f"Node {i}" for i in range(10)]
g2.vs["color"] = ["skyblue" for i in range(10)]


# Añadir aristas entre los nodos
# Opcion 1:
g2.add_edges([(0,1), (0,2), (0,3), (0,4), (1,2), (1,3), (1,4), (2,3), (2,4), (3,4), (5,6), (5,7), (5,8), (5,9), (6,7), (6,8), (6,9), (7,8), (7,9), (8,9), (4,5)])
# Opcion 2: crear aristas aleatorias entre nodos diferentes (cada vez que se ejecute el código, las aristas serán diferentes)
#edges = [tuple(random.sample(range(10), 2)) for i in range(10)]
#g2.add_edges(edges)

# Añadir atributos a las aristas (peso)
g2.es["weight"] = [1] * g2.ecount() # Peso inicial de las aristas (1 para todas) 
# Asignar "flujo" como letras a cada arista
g2.es["flow"] = [f"F{e.source},{e.target}" for e in g2.es]


# Mostrar el grafo
ig.plot(g2, vertex_label=g2.vs["name"], vertex_color=g2.vs["color"])
```

```python
# primero accedo al nodo/arista y luego al atributo
g2.vs[1:3]["name"] = ["A", "B"] 
g2.vs[3]["name"] = "C"   

g2.es[0:2]["weight"] = [5, 3]
ig.plot(g2, vertex_label=g2.vs["name"], vertex_color=g2.vs["color"], edge_width = g2.es["weight"], edge_label = g2.es["flow"])
```

```python
ig.plot(g2.as_directed())  # Convierte el grafo en no dirigido (si no lo es ya)
```

```python
print(g2.get_shortest_paths(0, 6))
```

```python
g2.delete_vertices(9)  # Elimina el nodo 9 y todas sus aristas asociadas
g2.delete_edges([(0,1), (2,3)])  # Elimina las aristas entre los nodos 0-1 y 2-3

ig.plot(g2, vertex_label=g2.vs["name"], vertex_color=g2.vs["color"], edge_width = g2.es["weight"], edge_label = g2.es["weight"])
```

```python
layout = g2.layout("fr")
ig.plot(g2, vertex_label=g2.vs["name"], vertex_color=g2.vs["color"], edge_width = g2.es["weight"], edge_label = g2.es["weight"], layout = layout)
```

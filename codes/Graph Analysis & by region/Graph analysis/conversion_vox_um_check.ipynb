## CHECKING VOX TO UM CONVERSION


```python
import pickle
import numpy as np

vox_pkl = "/home/admin/Ana/MicroBrain/output/vox/graph_18_OutGeom_Hcut1.pkl"
um_pkl  = "/home/admin/Ana/MicroBrain/output/um/graph_18_OutGeom_Hcut1_um.pkl"

def stats(arr, name):
    arr = np.asarray(arr)
    print(f"{name:16s}  min={arr.min():.3f}  max={arr.max():.3f}  shape={arr.shape}  dtype={arr.dtype}")

# --- VOXEL PKL ---
data_vox = pickle.load(open(vox_pkl, "rb"))
g_vox = data_vox["geom"]
v_vox = data_vox["vertex"]

print("\n=== VOX PKL ===")
stats(g_vox["x"], "geom[x]")
stats(g_vox["y"], "geom[y]")
stats(g_vox["z"], "geom[z]")
if "coords_image" in v_vox:
    stats(v_vox["coords_image"][:,0], "coords_img[x]")
    stats(v_vox["coords_image"][:,1], "coords_img[y]")
    stats(v_vox["coords_image"][:,2], "coords_img[z]")

# --- UM PKL ---
data_um = pickle.load(open(um_pkl, "rb"))
g_umR = data_um.get("geom_R", {})
v_umR = data_um.get("vertex_R", {})

print("\n=== UM PKL (geom_R / vertex_R) ===")
stats(g_umR["x_R"], "geom_R[x_R]")
stats(g_umR["y_R"], "geom_R[y_R]")
stats(g_umR["z_R"], "geom_R[z_R]")
if "coords_image_R" in v_umR:
    stats(v_umR["coords_image_R"][:,0], "coords_R[x]")
    stats(v_umR["coords_image_R"][:,1], "coords_R[y]")
    stats(v_umR["coords_image_R"][:,2], "coords_R[z]")

# --- OPTIONAL: check ratio (rough sanity) ---
# Only meaningful if ranges are large and not dominated by cropping.
print("\n=== Rough scale ratio (max) ===")
print("x ratio:", float(np.max(g_umR["x_R"])) / float(np.max(g_vox["x"])))
print("y ratio:", float(np.max(g_umR["y_R"])) / float(np.max(g_vox["y"])))
print("z ratio:", float(np.max(g_umR["z_R"])) / float(np.max(g_vox["z"])))
```

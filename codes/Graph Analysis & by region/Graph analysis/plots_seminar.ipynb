```python
import numpy as np
import matplotlib.pyplot as plt

# =========================
# DATOS
# =========================
boxes = ["HPC_1", "HPC_2", "HPC_3", "SMC_1", "SMC_2", "SMC_3"]

artery    = [96, 40, 72, 30, 35, 28]
vein      = [22, 26, 0, 66, 48, 35]
capillary = [1191, 1081, 810, 1409, 1594, 1385]

# =========================
# PLOT
# =========================
x = np.arange(len(boxes))
w = 0.25

fig, ax = plt.subplots(figsize=(9, 5))

bars1 = ax.bar(x - w, artery,    width=w, color="red",  label="Artery")
bars2 = ax.bar(x,     vein,      width=w, color="blue", label="Vein")
bars3 = ax.bar(x + w, capillary, width=w, color="gray", label="Capillary")

ax.set_xticks(x)
ax.set_xticklabels(boxes)
ax.set_xlabel("Box")
ax.set_ylabel("Value")
ax.set_title("Edge vessel type values by box")
ax.legend()
#ax.grid(axis="y", linestyle="--", alpha=0.4)

# Espacio extra arriba para que quepan las etiquetas
ax.set_ylim(0, 1750)

# Añadir valores encima de cada barra
ax.bar_label(bars1, padding=3, fontsize=10)
ax.bar_label(bars2, padding=3, fontsize=10)
ax.bar_label(bars3, padding=3, fontsize=10)

plt.tight_layout()

# Guardar figura
plt.savefig("./out/artery_vein_capillary_by_box.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# =========================
# DATOS
# =========================
boxes = ["HPC_1", "HPC_2", "HPC_3", "SMC_1", "SMC_2", "SMC_3"]

nodes = [1088, 947, 795, 1230, 1368, 1192]
edges = [1309, 1147, 912, 1505, 1677, 1448]

# =========================
# PLOT
# =========================
x = np.arange(len(boxes))
w = 0.35

fig, ax = plt.subplots(figsize=(10, 5))

bars1 = ax.bar(x - w/2, nodes, width=w, label="Nodes", color="lightcoral")
bars2 = ax.bar(x + w/2, edges, width=w, label="Edges", color = "lightsteelblue")

ax.set_xticks(x)
ax.set_xticklabels(boxes)
ax.set_xlabel("Box")
ax.set_ylabel("Count")
ax.set_title("Nodes and Edges by box")
ax.legend()
#ax.grid(axis="y", linestyle="--", alpha=0.4)

# etiquetas encima de las barras
ax.bar_label(bars1, padding=3, fontsize=10)
ax.bar_label(bars2, padding=3, fontsize=10)

plt.tight_layout()

# guardar figura
plt.savefig("./out/nodes_edges_by_box.png", dpi=300, bbox_inches="tight")

plt.show()
```

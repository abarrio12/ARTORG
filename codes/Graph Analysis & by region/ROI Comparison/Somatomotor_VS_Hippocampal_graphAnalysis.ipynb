```python
import numpy as np
import matplotlib.pyplot as plt

import graph_analysis_functions as gaf

# Importar los notebooks como módulos
from ipynb.fs.full import HippocampalArea_graphAnalysis as hippo
from ipynb.fs.full import SomatomotorArea_graphAnalysis as soma


vessel_type = {
    2: "arteriole",
    3: "venule",
    4: "capillary"
}

# Hippocampal
hip_data   = hippo.data
hip_nkind  = hippo.nkind
hip_diam   = hippo.diameters_hippo
hip_len    = hippo.lengths_hippo


# Somatomotor
so_data    = soma.data
so_nkind   = soma.nkind
so_diam    = soma.diameters_soma
so_len     = soma.lengths_soma
```

```python
plt.figure(figsize=(8,5))

x = np.arange(len(hip_nkind))

plt.bar(x - 0.2, hip_diam, width=0.4, label="Hippocampal")
plt.bar(x + 0.2, so_diam,  width=0.4, label="Somatomotor")

plt.xticks(x, [hippo.vessel_type[k] for k in hip_nkind])
plt.ylabel("Diameter (µm)")
plt.title("Diameter comparison by vessel type")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))

x = np.arange(len(hip_nkind))

plt.bar(x - 0.2, hip_len, width=0.4, label="Hippocampal")
plt.bar(x + 0.2, so_len,  width=0.4, label="Somatomotor")

plt.xticks(x, [hippo.vessel_type[k] for k in hip_nkind])
plt.ylabel("Length (µm)")
plt.title("Length comparison by vessel type")
plt.legend()
plt.show()
```

```python
# xmin and xmax of all distributions
global_min = min(hippo.diameters_dist_hippo.min(), soma.diameters_dist_soma.min())
global_max = max(hippo.diameters_dist_hippo.max(), soma.diameters_dist_soma.max())

unique_kinds = np.unique(hippo.nkind_distribution)

for k in unique_kinds:
    plt.figure(figsize=(8,5))

    # Filter data by nkind
    hip = hippo.diameters_dist_hippo[hippo.nkind_distribution == k]
    so  = soma.diameters_dist_soma[soma.nkind_distribution == k]


    plt.hist(hip, bins=80, alpha=0.6, density=True, label="Hippocampal")
    plt.hist(so,  bins=80, alpha=0.6, density=True, label="Somatomotor")

    # same range for all plots
    plt.xlim(global_min, global_max)

    # global y 
    hip_counts, _ = np.histogram(hip, bins=80, density=True)
    so_counts,  _ = np.histogram(so,  bins=80, density=True)
    ymax = max(hip_counts.max(), so_counts.max()) * 1.15 # gets max of all histograms and *1.15 for visual margin
    plt.ylim(0, ymax)

    # -------------------------------

    plt.title(f"Diameter distribution comparison – nkind = {k} ({hippo.vessel_type[k]})")
    plt.xlabel("Diameter (µm)")
    plt.ylabel("Density")
    plt.legend()
    plt.tight_layout()
    plt.show()
```
